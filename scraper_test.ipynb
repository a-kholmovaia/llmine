```python
import requests
from bs4 import BeautifulSoup

def scrape_links(url, links):
    try:
        # Send an HTTP GET request to the specified URL
        response = requests.get(url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the HTML content of the page using BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')

            # Find all the anchor (a) tags in the body of the HTML
            links = soup.find_all('a')

            # Extract and print the href attribute from each anchor tag
            for link in links:
                href = link.get('href')
                if href:
                    links.append
        else:
            print(f"Error: Unable to fetch the webpage. Status code: {response.status_code}")

    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
url_to_scrape = "https://minecraft.fandom.com/wiki/Item"
scrape_links(url_to_scrape)
```
